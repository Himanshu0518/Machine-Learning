```python
# plotting a parabola
x = np.linspace(-10,10,100)
y = x*x ;
plt.plot(x,y)
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.title('parabolic graph')
plt.show()
```

**Bar Plot**

```python
# way1
languages = ['English','Hindi','French','Germany']
people = [100,50,40,45]
colors = ['red','green','yellow','blue']
plt.bar(languages,people,color=colors)
plt.show()
```

```python
plt.barh(languages,people,color=colors)
plt.show()
```

```python
fig =  plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','Hindi','French','Germany']
people = [100,50,40,45]
ax.bar(languages,people)
plt.xlabel('Number of People')
plt.ylabel('Languages')
```

**Pie Plot**

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels,shadow='true',explode=myexplode)
plt.show()
```

we can also mention the start angle . By default x axis is the start angle


```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
languages = ['English','Hindi','French','Germany']
people = [100,50,40,45]
ax1.pie(people,labels=languages,autopct='%1.1f%%',startangle=90)
plt.legend(title='languages')
plt.show()
```

**Scatter plot**


```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

**Markers in Matplot**

ms : use to   mention the size of marker

mec : use to mention the color of marker

mfc : use to set marker face color

```python
y = np.array([3,5,7,9,4,6,1]) # when x is not given then indices are used as x values
plt.plot(y,marker='o',ms=10,mec='r',mfc='g')
```

Shortcut :- format strings fmt

'-'	Solid line

':'	Dotted line

'--'	Dashed line

'-.'	Dashed/dotted line

```python
plt.plot(y,'*:r')
```

**Adding grid lines**

```python
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)  # color,linestyle and linewidth are opptional
plt.plot(y,'o:k',ms=5,mec='r')
plt.show()
```

**Subplot using Matplot**

```python
# Now let us plot some basic elementary signals using matplot

t = np.arange(-5,5,1) ;
# unit signal or heviside function
ut = np.where(t>=0,1,0)
plt.subplot(2,2,1)
plt.plot(t,ut)

plt.subplot(2,2,2)
# rectangular pulse signal
T = 4
rec = np.where((t>=-T/2) & (t<=T/2) ,1,0)
plt.plot(t,rec)

# unit ramp signal
rt = np.where(t>=0,t,0)
plt.subplot(2,2,3)
plt.plot(t,rt)

# unit imulse signal
impulse = np.where(t==0,1,0)
plt.subplot(2,2,4)
plt.plot(t,impulse)
```

Using a Color Map in Matplot

This colormap is called 'viridis' and as you can see it ranges from 0, which is a purple color, up to 100, which is a yellow color.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python

x = np.random.randint(100,size=(100))
y = np.random.randint(100,size=(100))

colors = np.random.randint(100,size=(100))
sizes = 10*np.random.randint(100, size=(100))

plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='nipy_spectral')

plt.colorbar()
plt.show()
```
