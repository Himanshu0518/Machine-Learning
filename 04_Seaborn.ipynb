#**Seaborn**

Used for data visualisation

```python
# Importing the seaborn library
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Displots**

Distplot stands for distribution plot, it takes as input an array and plots a curve corresponding to the distribution of points in the array.

```python
sns.distplot([0, 1, 2, 3, 4, 5])
plt.show()
```

```python
# without using histogram
sns.distplot([0, 1, 2, 3, 4, 5], hist=False)
plt.show()
```

Generate a random normal distribution

```python
from numpy import random
x = random.normal(size=(2,3),loc=2,scale=3)
print(x)
```

```python
mean = 2
st_dev = 3
sns.distplot(random.normal(loc=mean,scale=st_dev,size=1000), hist=False )
```

**Binomial Distribution**

```python
b = random.binomial(n=10,p=0.5,size=15)
print(b)
```

```python
sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True,kde=False)
plt.show()
```

```python

sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='binomial')
plt.show()
```

**Poisson's distribution**

```python
x = random.poisson(lam=2 , size=10)
print(x)
print("Mean:",x.mean())
print("Standard-deviation :",x.std())
```

Visualisation of poisson distribution

```python
sns.displot(random.poisson(lam=2,size = 100),kde=False)
```

**Uniform Distribution**

Used to describe probability where every event has equal chances of occuring.

* It is continuous probability distribution

---



```python
sns.distplot(random.uniform(size=1000), hist=False,bins=20)
plt.show()
```

```python
# house price dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
housing_df = pd.DataFrame(housing.data,columns=housing.feature_names)
housing_df['PRICE'] = housing.target
```

```python
housing_df.head()
```

Distribution Plot

```python
sns.displot(housing_df['PRICE'])
```

```python
sns.distplot(housing_df['PRICE'])
```

Correlation :
1. positive Correlation
2. negative correlation

Heat Map

```python
correlation = housing_df.corr() ;
```

```python
# Constructing a heat Map
plt.figure()
sns.heatmap(correlation , cbar=True , square=True,fmt='.1f',annot=True,cmap='Blues')
```
