```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
```

```python
dataset = datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(dataset.data,columns = dataset.feature_names)
df.head()
```

```python

X = df
Y = dataset.target
```

```python
df['target'] = Y
```

```python
print(X[0:1])
```

```python
print(Y)
```

1--> Benign

0-->Malingnant

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

#GridSearchCV

```python
model = SVC()
```

```python
parameters = {
    'kernel' : ['linear','poly','rbf','sigmoid'],
    'C':[1 , 5 ,10 , 20]
}
```

```python
classifier = GridSearchCV(model,parameters,cv=5)
```

```python
classifier.fit(X,Y)
```

```python
pd.DataFrame(classifier.cv_results_).head()
```

```python
classifier.best_params_
```

```python
classifier.best_score_
```

#RandomisedSearchCV

```python
model = SVC()
```

```python
parameters = {
    'kernel' : ['linear','poly','rbf','sigmoid'],
    'C':[1 , 5 ,10 , 20]
}
```

```python
classifier = RandomizedSearchCV(model,parameters,cv=5)
```

```python
classifier.fit(X,Y)
```

```python
classifier.best_params_
```

```python
classifier.best_score_
```

```python
result = classifier.cv_results_
```

```python
res = pd.DataFrame(result)
```

```python
res.head()
```

```python
res = res[['param_kernel','param_C','mean_test_score']]
```
